# Why this data

Using the best neural network player that our Alpha Awari was able to produce, three datasets have been created in order to better analyze the strenght and understand how many times the network is able to choose the correct move. In this work an analysis on the three datasets is provided.

The three datasets are: 
-	Matches dataset, dataset that describe how many optimal and suboptimal moves the network can choose on each match.
-	Moves dataset, dataset that describe in deep the moves of our work.
-   Opponent dataset, summarize the number of wins or lose for each different opponent.

# How many optimal moves choose the network?

```python
optimal_moves = moves_data[moves_data['best_move_selected']==1]
optimal_moves = optimal_moves['best_move_selected'].count()
optimal_moves
```

```python
soptimal_moves = moves_data[moves_data['best_move_selected']==0]
soptimal_moves = soptimal_moves['best_move_selected'].count()
soptimal_moves
```

```python
def annotate_plot(plot, x=.18, y=.2, font=15):
    # create a list to collect the plt.patches data
    totals = []

    # find the values and append to list
    for i in plot.patches:
        totals.append(i.get_height())

    # set individual bar lables using above list
    total = sum(totals)

    # set individual bar lables using above list
    for i in plot.patches:
        # get_x pulls left or right; get_height pushes up or down
        plot.text(i.get_x()+x, i.get_height()+y, str(round((i.get_height()/total)*100, 2))+'%', fontsize=font, color='dimgrey')
```

```python
optimal_moves_bar = pd.Series({'Optimal moves': optimal_moves,
               'suboptimal' : soptimal_moves})

plot = optimal_moves_bar.plot.bar(figsize=(10,8), width=0.3)

annotate_plot(plot)
plt.title("Optimal moves chosen by Neural Network")
plt.xticks(rotation=0)
plt.show()
```

```python
wins = matches_data[matches_data['win']==1]
wins = wins['win'].count()
wins
```

```python
lost = matches_data[matches_data['win']==0]
lost = lost['win'].count()
lost
```

```python
win_rate = pd.Series({'Win': wins,
               'Lost' : lost})

plot = win_rate.plot.bar(figsize=(10,8), width=0.3)

annotate_plot(plot)
plt.title('Win rate')
plt.ylabel('Number of wins')
plt.xticks(rotation=0)
plt.show()
```
